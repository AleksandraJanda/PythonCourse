```python
df3 = df.drop(['CCA3','Country','Capital'], axis= 1)
list_columns = ['2022 Population', '2020 Population',
       '2015 Population', '2010 Population', '2000 Population',
       '1990 Population', '1980 Population', '1970 Population']
df3 = df3.groupby('Continent')[list_columns[::-1]].mean().sort_values(by= '2022 Population', ascending= False)
df3
```

```python
new_columns = df3.columns.str.replace(' Population', '')
df3.columns = new_columns
df3
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df4 = df3.transpose()
df4
```
